```python
import pandas as pd
import matplotlib.pyplot as plt

file="/kaggle/input/imdb-movies-data/imdb-movies-dataset.csv"
imdb_data=pd.read_csv(file)
```

```python
imdb_data=imdb_data[["Year","Rating"]].dropna()

# avg rating per year
rating_trends=imdb_data.groupby("Year").mean() 

plt.plot(rating_trends.index,rating_trends["Rating"],'g',marker='.',linestyle='-')
plt.title("Evolution of ratings(IMDB)")
plt.xlabel("Year")
plt.ylabel("Rating IMDB")
plt.grid()
plt.show()
```
