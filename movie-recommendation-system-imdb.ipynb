```python
import pandas as pd
file="/kaggle/input/imdb-movies-data/imdb-movies-dataset.csv"
imdb_data=pd.read_csv(file)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer

imdb_data=imdb_data[["Title","Genre","Director","Cast"]].dropna()
imdb_data["features"]=imdb_data["Genre"]+" "+imdb_data["Director"]+" "+imdb_data["Cast"]

vectorizer = CountVectorizer().fit_transform(imdb_data["features"])

similarity_matrix = cosine_similarity(vectorizer)
def recommend_movies(movie_name):
    if movie_name not in imdb_data["Title"].values:
        print("The movie does not exist in the dataset!")
        return
    idx=imdb_data[imdb_data["Title"]==movie_name].index[0]
    scores=list(enumerate(similarity_matrix[idx]))
    sorted_movies = sorted(scores, key=lambda x: x[1], reverse=True)

    print(f"Recommended for '{movie_name}':")
    for i in range(1, 10):
        movie_idx = sorted_movies[i][0]
        print(f"- {imdb_data.iloc[movie_idx]['Title']}")
```

```python
recommend_movies("Inception")
```

```python
recommend_movies("Despicable Me 4")
```
